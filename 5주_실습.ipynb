```python
### 발행년도가 없는 값을 뽑아 df로 만드시오
import numpy as np
no_year = df[df['발행년도'].isnull()]
no_year.head()
```

```python
#단, 출판사가 없을 시 '출판사 없음' 입력 후 df 확인
no_year = df[df['발행년도'].isnull()]
no_year['출판사'].fillna('출판사 없음', inplace=True)
no_year.head()
```

```python
#발행년도가 없는 출판사를 찾아, 출판사별 '발행년도 최빈값'으로 채워넣으시오
#힌트! 발행년도가 없는 출판사의 리스트를 뽑아 for loop처리!
no_year = df[df['발행년도'].isnull()]['출판사'].unique()
for publisher in no_year:
    years = df[(df['출판사'] == publisher) & (~df['발행년도'].isnull())]['발행년도']
    if not years.empty:
        mode_year = years.mode().values[0]
        df.loc[(df['출판사'] == publisher) & (df['발행년도'].isnull()), '발행년도'] = mode_year
df
```
